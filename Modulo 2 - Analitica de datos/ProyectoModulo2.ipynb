# Despues de conocer y analizar la etapa de diagnostico, ahora sigue hacer la limpieza de datos con las instrucciones dadas en nuestro Proyecto


```python
#Conteo de registros faltantes por fecha
columnas = ["PRECIP", "EVAP", "TMAX", "TIMN"]
df["faltantes"] = df[columnas].isnull().sum(axis=1)
df[df["faltantes"] > 0][["FECHA", "faltantes"]]

#La respuesta indicara cuantos nulos hay faltantes en cada fecha, dando un total de 312 columnas con registros faltantes
```

```python
# Rellenar nulos con la media de la columna
df["PRECIP"] = df["PRECIP"].fillna(df["PRECIP"].mean())
df["EVAP"] = df["EVAP"].fillna(df['EVAP'].mean(),inplace=True)
df["TMAX"] = df["TMAX"].fillna(df['TMAX'].mean(),inplace=True)
df["TIMN"] = df["TIMN"].fillna(df['TIMN'].mean(),inplace=True)
```

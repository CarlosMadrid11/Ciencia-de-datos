````python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# IMPORTACIONES
# Solo usamos 'date' de la librerÃ­a estÃ¡ndar de Python para registrar
# automÃ¡ticamente la fecha del dÃ­a en que se realiza cada prÃ©stamo.
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
from datetime import date
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
#  SECCIÃ“N 1 â€” CLASES (POO)
#  Se definen tres clases que representan las entidades del sistema.
#  Cada clase agrupa sus propios datos (atributos) y acciones (mÃ©todos),
#  lo que hace el cÃ³digo mÃ¡s organizado y fÃ¡cil de mantener.
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•


class Libro:
    """
    Representa un libro dentro del catÃ¡logo de la biblioteca.

    Atributos:
        isbn (str)                  : Identificador Ãºnico del libro.
        titulo (str)                : TÃ­tulo del libro.
        autor (str)                 : Nombre del autor.
        ejemplares_totales (int)    : Cantidad total de copias del libro.
        ejemplares_disponibles (int): Copias disponibles para prÃ©stamo.
    """

    def __init__(self, isbn, titulo, autor, ejemplares_totales):
        self.isbn = isbn
        self.titulo = titulo
        self.autor = autor
        self.ejemplares_totales = ejemplares_totales
        # Al registrar el libro, todos los ejemplares estÃ¡n disponibles
        self.ejemplares_disponibles = ejemplares_totales

    def mostrar(self):
        """
        Devuelve un texto legible con la informaciÃ³n del libro.
        Ejemplo: '[ISBN: 978-1] Cien aÃ±os de soledad â€” GarcÃ­a MÃ¡rquez | Disponibles: 2/3'
        """
        return (
            f"[ISBN: {self.isbn}] '{self.titulo}' â€” {self.autor} "
            f"| Disponibles: {self.ejemplares_disponibles}/{self.ejemplares_totales}"
        )


class Usuario:
    """
    Representa a una persona registrada en la biblioteca.

    Atributos:
        id_usuario (str)       : Identificador Ãºnico del usuario.
        nombre (str)           : Nombre completo del usuario.
        prestamos_activos (int): Cantidad de libros que tiene prestados actualmente.
                                 Sirve para controlar el lÃ­mite de 3 prÃ©stamos (RF5).
    """

    LIMITE_PRESTAMOS = 3  # Constante de clase: mÃ¡ximo prÃ©stamos permitidos por usuario

    def __init__(self, id_usuario, nombre):
        self.id_usuario = id_usuario
        self.nombre = nombre
        self.prestamos_activos = 0  # Empieza en 0, sin prÃ©stamos

    def puede_pedir_prestado(self):
        """
        Retorna True si el usuario aÃºn puede pedir libros prestados,
        False si ya alcanzÃ³ el lÃ­mite de 3 prÃ©stamos activos.
        """
        return self.prestamos_activos < self.LIMITE_PRESTAMOS

    def mostrar(self):
        """
        Devuelve un texto legible con la informaciÃ³n del usuario.
        Ejemplo: '[ID: U01] Ana LÃ³pez | PrÃ©stamos activos: 2/3'
        """
        return (
            f"[ID: {self.id_usuario}] {self.nombre} "
            f"| PrÃ©stamos activos: {self.prestamos_activos}/{self.LIMITE_PRESTAMOS}"
        )


class Biblioteca:
    """
    Clase principal que gestiona todas las operaciones del sistema.

    Contiene y administra:
        catalogo (dict)             : Diccionario {isbn: Libro}.
                                      Permite buscar cualquier libro por ISBN en O(1).
        usuarios (dict)             : Diccionario {id_usuario: Usuario}.
                                      Permite buscar cualquier usuario por ID en O(1).
        prestamos (list)            : Lista de tuplas (isbn, id_usuario, fecha).
                                      Cada tupla es un registro inmutable de un prÃ©stamo activo.
        frecuencias_prestamos (dict): Diccionario {isbn: total_prestamos_historicos}.
                                      Lleva la cuenta total de veces que se prestÃ³ cada libro
                                      para generar el reporte Top 3 (RF4).
    """

    def __init__(self):
        # DICCIONARIO: acceso rÃ¡pido por ISBN para el catÃ¡logo de libros
        self.catalogo: dict = {}

        # DICCIONARIO: acceso rÃ¡pido por ID para los usuarios registrados
        self.usuarios: dict = {}

        # LISTA: colecciÃ³n ordenada de prÃ©stamos activos (cada elemento es una TUPLA)
        self.prestamos: list = []

        # DICCIONARIO: contador histÃ³rico de prÃ©stamos por ISBN (para reportes)
        self.frecuencias_prestamos: dict = {}

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # RF1 â€” GESTIÃ“N DE LIBROS
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

    def agregar_libro(self, isbn, titulo, autor, ejemplares_totales):
        """
        Registra un nuevo libro en el catÃ¡logo.
        Si ya existe un libro con ese ISBN, informa al usuario y no hace nada.
        """
        # Verificar que no exista un libro con el mismo ISBN
        if isbn in self.catalogo:
            print("  âš  Ya existe un libro con ese ISBN en el catÃ¡logo.")
            return

        # Validar que el nÃºmero de ejemplares sea positivo
        if ejemplares_totales <= 0:
            print("  âš  El nÃºmero de ejemplares debe ser mayor a 0.")
            return

        # Crear el objeto Libro y guardarlo en el diccionario catalogo
        self.catalogo[isbn] = Libro(isbn, titulo, autor, ejemplares_totales)
        print(f"  âœ” Libro '{titulo}' agregado correctamente.")

    def consultar_disponibilidad(self, isbn):
        """
        Muestra cuÃ¡ntos ejemplares de un libro estÃ¡n disponibles para prÃ©stamo.
        Busca el libro por ISBN en el diccionario catalogo.
        """
        # El mÃ©todo .get() devuelve None si la clave no existe (sin lanzar error)
        libro = self.catalogo.get(isbn)
        if libro:
            print(f"  {libro.mostrar()}")
        else:
            print("  âš  No se encontrÃ³ ningÃºn libro con ese ISBN.")

    def buscar_libro(self, termino):
        """
        Busca libros cuyo tÃ­tulo o autor contengan el tÃ©rmino ingresado.
        La bÃºsqueda no distingue entre mayÃºsculas y minÃºsculas.
        Recorre todos los valores del diccionario catalogo.
        """
        termino = termino.lower()  # Convertir a minÃºsculas para comparaciÃ³n flexible

        # Usar comprensiÃ³n de lista para filtrar libros que coincidan
        encontrados = [
            libro for libro in self.catalogo.values()
            if termino in libro.titulo.lower() or termino in libro.autor.lower()
        ]

        if encontrados:
            print(f"  Se encontraron {len(encontrados)} libro(s):")
            for libro in encontrados:
                print(f"    â†’ {libro.mostrar()}")
        else:
            print(f"  âš  No se encontraron libros con el tÃ©rmino '{termino}'.")

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # RF2 â€” REGISTRO DE USUARIOS
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

    def agregar_usuario(self, id_usuario, nombre):
        """
        Registra un nuevo usuario en el sistema.
        El ID de usuario debe ser Ãºnico; si ya existe, no se permite duplicarlo.
        """
        if id_usuario in self.usuarios:
            print("  âš  Ya existe un usuario con ese ID.")
            return

        if not nombre.strip():
            print("  âš  El nombre del usuario no puede estar vacÃ­o.")
            return

        # Crear el objeto Usuario y guardarlo en el diccionario usuarios
        self.usuarios[id_usuario] = Usuario(id_usuario, nombre.strip())
        print(f"  âœ” Usuario '{nombre}' registrado correctamente.")

    def consultar_usuarios(self):
        """
        Muestra todos los usuarios registrados con su informaciÃ³n actual.
        """
        if not self.usuarios:
            print("  âš  No hay usuarios registrados todavÃ­a.")
            return

        print(f"  Usuarios registrados ({len(self.usuarios)} en total):")
        for usuario in self.usuarios.values():
            print(f"    â†’ {usuario.mostrar()}")

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # RF3 â€” PRÃ‰STAMOS Y DEVOLUCIONES
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

    def prestar_libro(self, isbn, id_usuario):
        """
        Registra el prÃ©stamo de un libro a un usuario.

        Antes de prestar, se verifican todas las reglas de negocio (RF5):
          1. El libro debe existir en el catÃ¡logo.
          2. El usuario debe estar registrado.
          3. Debe haber al menos un ejemplar disponible.
          4. El usuario no debe tener ya ese mismo libro prestado (sin duplicados).
          5. El usuario no debe superar el lÃ­mite de 3 prÃ©stamos activos.

        Si todo es correcto:
          - Se crea una TUPLA inmutable (isbn, id_usuario, fecha) como registro del prÃ©stamo.
          - La tupla se agrega a la LISTA de prÃ©stamos activos.
          - Se actualiza el contador del usuario y los ejemplares disponibles del libro.
          - Se actualiza el DICCIONARIO de frecuencias para el reporte Top 3.
        """
        # Regla 1: Verificar que el libro exista
        if isbn not in self.catalogo:
            print("  âš  No existe ningÃºn libro con ese ISBN.")
            return

        # Regla 2: Verificar que el usuario estÃ© registrado
        if id_usuario not in self.usuarios:
            print("  âš  No existe ningÃºn usuario con ese ID.")
            return

        # Obtener los objetos Libro y Usuario para trabajar con ellos
        libro = self.catalogo[isbn]
        usuario = self.usuarios[id_usuario]

        # Regla 3: Verificar que haya ejemplares disponibles
        if libro.ejemplares_disponibles <= 0:
            print(f"  âš  No hay ejemplares disponibles de '{libro.titulo}'.")
            return

        # Regla 4: Verificar que el usuario no tenga ya ese libro prestado
        # Recorre la lista de tuplas y compara isbn e id_usuario
        for prestamo in self.prestamos:
            if prestamo[0] == isbn and prestamo[1] == id_usuario:
                print(f"  âš  {usuario.nombre} ya tiene prestado '{libro.titulo}'.")
                return

        # Regla 5: Verificar que el usuario no haya superado el lÃ­mite de prÃ©stamos
        if not usuario.puede_pedir_prestado():
            print(f"  âš  {usuario.nombre} ya tiene {usuario.LIMITE_PRESTAMOS} prÃ©stamos activos (lÃ­mite mÃ¡ximo).")
            return

        # â”€â”€ Todas las validaciones pasaron, registrar el prÃ©stamo â”€â”€

        # TUPLA: registro inmutable del prÃ©stamo (no puede modificarse accidentalmente)
        nuevo_prestamo = (isbn, id_usuario, date.today())

        # LISTA: agregar la tupla a la colecciÃ³n de prÃ©stamos activos
        self.prestamos.append(nuevo_prestamo)

        # Actualizar disponibilidad del libro y contador del usuario
        libro.ejemplares_disponibles -= 1
        usuario.prestamos_activos += 1

        # DICCIONARIO: actualizar frecuencias histÃ³ricas (para el reporte Top 3)
        # .get(isbn, 0) devuelve 0 si el ISBN no existe aÃºn en el diccionario
        self.frecuencias_prestamos[isbn] = self.frecuencias_prestamos.get(isbn, 0) + 1

        print(f"  âœ” PrÃ©stamo registrado exitosamente.")
        print(f"     Libro   : '{libro.titulo}'")
        print(f"     Usuario : {usuario.nombre}")
        print(f"     Fecha   : {nuevo_prestamo[2]}")

    def devolver_libro(self, isbn, id_usuario):
        """
        Registra la devoluciÃ³n de un libro.

        Busca en la LISTA de prÃ©stamos la TUPLA que corresponde a ese libro y usuario.
        Si la encuentra:
          - Elimina la tupla de la lista de prÃ©stamos activos.
          - Libera el ejemplar (suma 1 a ejemplares_disponibles del libro).
          - Resta 1 al contador de prÃ©stamos activos del usuario.
        """
        # Verificar que el libro y usuario existan antes de buscar
        if isbn not in self.catalogo:
            print("  âš  No existe ningÃºn libro con ese ISBN.")
            return
        if id_usuario not in self.usuarios:
            print("  âš  No existe ningÃºn usuario con ese ID.")
            return

        # Buscar el prÃ©stamo activo en la lista de tuplas
        for prestamo in self.prestamos:
            if prestamo[0] == isbn and prestamo[1] == id_usuario:
                # Encontrado: eliminar de la lista y liberar el ejemplar
                self.prestamos.remove(prestamo)
                self.catalogo[isbn].ejemplares_disponibles += 1
                self.usuarios[id_usuario].prestamos_activos -= 1
                print(f"  âœ” DevoluciÃ³n registrada: '{self.catalogo[isbn].titulo}'")
                print(f"     Devuelto por: {self.usuarios[id_usuario].nombre}")
                return

        # Si el bucle termina sin encontrar el prÃ©stamo:
        print("  âš  No se encontrÃ³ ese prÃ©stamo activo.")

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # RF4 â€” REPORTES Y CONSULTAS
    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

    def top3_libros_mas_prestados(self):
        """
        Muestra el Top 3 de los libros mÃ¡s solicitados en la historia.

        Usa el DICCIONARIO de frecuencias (frecuencias_prestamos) que se va
        actualizando con cada prÃ©stamo. Lo ordena por valor (cantidad de prÃ©stamos)
        de mayor a menor y toma los primeros 3 resultados.
        """
        if not self.frecuencias_prestamos:
            print("  âš  AÃºn no se ha registrado ningÃºn prÃ©stamo.")
            return

        # sorted() ordena el diccionario como lista de tuplas (isbn, cantidad)
        # key=lambda x: x[1] indica que se ordene por el segundo elemento (cantidad)
        # reverse=True indica orden descendente (de mayor a menor)
        top3 = sorted(
            self.frecuencias_prestamos.items(),
            key=lambda x: x[1],
            reverse=True
        )[:3]  # [:3] toma solo los primeros 3 resultados

        print("  ğŸ† Top 3 libros mÃ¡s prestados:")
        medallas = ["ğŸ¥‡", "ğŸ¥ˆ", "ğŸ¥‰"]
        for i, (isbn, cantidad) in enumerate(top3):
            # Obtener el tÃ­tulo del libro desde el catÃ¡logo
            titulo = self.catalogo[isbn].titulo if isbn in self.catalogo else isbn
            print(f"     {medallas[i]} {titulo} â€” {cantidad} prÃ©stamo(s) en total")

    def listar_prestamos_activos(self):
        """
        Muestra todos los prÃ©stamos que estÃ¡n actualmente en curso.

        Recorre la LISTA de prÃ©stamos. Cada elemento es una TUPLA de 3 valores:
        (isbn, id_usuario, fecha). Se usa desempaquetado de tupla para leerlos.
        """
        if not self.prestamos:
            print("  âš  No hay prÃ©stamos activos en este momento.")
            return

        print(f"  PrÃ©stamos activos ({len(self.prestamos)} en total):")
        for isbn, id_usuario, fecha in self.prestamos:  # Desempaquetado de tupla
            titulo = self.catalogo[isbn].titulo if isbn in self.catalogo else isbn
            nombre = self.usuarios[id_usuario].nombre if id_usuario in self.usuarios else id_usuario
            print(f"    â†’ '{titulo}' | {nombre} | Desde: {fecha}")

    def libros_por_autor(self, autor):
        """
        Muestra todos los libros de un autor especÃ­fico.
        La bÃºsqueda no distingue mayÃºsculas y permite nombres parciales.
        """
        autor_lower = autor.lower()
        encontrados = [
            libro for libro in self.catalogo.values()
            if autor_lower in libro.autor.lower()
        ]

        if encontrados:
            print(f"  Libros del autor '{autor}' ({len(encontrados)} encontrado(s)):")
            for libro in encontrados:
                print(f"    â†’ {libro.mostrar()}")
        else:
            print(f"  âš  No se encontraron libros del autor '{autor}'.")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
#  SECCIÃ“N 2 â€” FUNCIONES UTILITARIAS
#  Funciones independientes que dan soporte a la interfaz de consola.
#  EstÃ¡n separadas de las clases para mantener la responsabilidad Ãºnica:
#  las clases manejan los datos, estas funciones manejan la presentaciÃ³n.
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•


def limpiar_pantalla():
    """Imprime lÃ­neas en blanco para simular limpieza de pantalla en consola."""
    print("\n" * 2)


def separador(caracter="â”€", longitud=50):
    """Imprime una lÃ­nea separadora para organizar visualmente la consola."""
    print("  " + caracter * longitud)


def pedir_entero(mensaje, minimo=1):
    """
    Solicita al usuario que ingrese un nÃºmero entero vÃ¡lido.
    Repite la pregunta hasta recibir un valor aceptable (>= minimo).
    Esta funciÃ³n evita que el programa se rompa si el usuario escribe letras.
    """
    while True:
        try:
            valor = int(input(mensaje))
            if valor >= minimo:
                return valor
            print(f"  âš  El nÃºmero debe ser al menos {minimo}. Intenta de nuevo.")
        except ValueError:
            print("  âš  Por favor ingresa un nÃºmero entero vÃ¡lido.")


def pedir_texto(mensaje):
    """
    Solicita al usuario que ingrese texto no vacÃ­o.
    Repite la pregunta si el usuario presiona Enter sin escribir nada.
    """
    while True:
        texto = input(mensaje).strip()
        if texto:
            return texto
        print("  âš  Este campo no puede estar vacÃ­o. Intenta de nuevo.")


def mostrar_menu_principal():
    """Imprime el menÃº principal del sistema con todas las opciones disponibles."""
    print("""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘      ğŸ“š SISTEMA DE BIBLIOTECA ESCOLAR    â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘                                          â•‘
â•‘   â”€â”€ LIBROS â”€â”€                           â•‘
â•‘    1. Agregar libro al catÃ¡logo          â•‘
â•‘    2. Consultar disponibilidad de libro  â•‘
â•‘    3. Buscar libro por tÃ­tulo o autor    â•‘
â•‘    4. Ver libros de un autor             â•‘
â•‘                                          â•‘
â•‘   â”€â”€ USUARIOS â”€â”€                         â•‘
â•‘    5. Registrar nuevo usuario            â•‘
â•‘    6. Ver todos los usuarios             â•‘
â•‘                                          â•‘
â•‘   â”€â”€ PRÃ‰STAMOS â”€â”€                        â•‘
â•‘    7. Prestar libro a usuario            â•‘
â•‘    8. Registrar devoluciÃ³n               â•‘
â•‘    9. Ver prÃ©stamos activos              â•‘
â•‘                                          â•‘
â•‘   â”€â”€ REPORTES â”€â”€                         â•‘
â•‘   10. Top 3 libros mÃ¡s prestados         â•‘
â•‘                                          â•‘
â•‘    0. Salir del sistema                  â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•""")


def encabezado(titulo):
    """Imprime un encabezado con el tÃ­tulo de la secciÃ³n activa."""
    limpiar_pantalla()
    separador("â•")
    print(f"  {titulo}")
    separador("â•")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
#  SECCIÃ“N 3 â€” FUNCIÃ“N PRINCIPAL (main)
#  Es el punto de entrada del sistema. Crea la biblioteca y gestiona
#  el flujo del menÃº en un bucle continuo hasta que el usuario decida salir.
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•


def main():
    """
    Inicia el sistema de biblioteca escolar.
    Mantiene un bucle principal que muestra el menÃº y ejecuta la opciÃ³n elegida.
    """
    # Crear la instancia principal de la biblioteca
    biblioteca = Biblioteca()

    print("  Bienvenido al Sistema de GestiÃ³n de Biblioteca Escolar.")
    print("  Escribe el nÃºmero de la opciÃ³n que deseas y presiona Enter.")

    while True:
        mostrar_menu_principal()
        opcion = input("  Â¿QuÃ© deseas hacer? Escribe el nÃºmero: ").strip()

        # â”€â”€ LIBROS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

        if opcion == "1":
            encabezado("AGREGAR LIBRO AL CATÃLOGO")
            isbn     = pedir_texto("  ISBN del libro     : ")
            titulo   = pedir_texto("  TÃ­tulo             : ")
            autor    = pedir_texto("  Autor              : ")
            ejemplares = pedir_entero("  NÃºmero de ejemplares: ", minimo=1)
            separador()
            biblioteca.agregar_libro(isbn, titulo, autor, ejemplares)

        elif opcion == "2":
            encabezado("CONSULTAR DISPONIBILIDAD")
            isbn = pedir_texto("  ISBN del libro: ")
            separador()
            biblioteca.consultar_disponibilidad(isbn)

        elif opcion == "3":
            encabezado("BUSCAR LIBRO")
            termino = pedir_texto("  Escribe tÃ­tulo o autor a buscar: ")
            separador()
            biblioteca.buscar_libro(termino)

        elif opcion == "4":
            encabezado("LIBROS POR AUTOR")
            autor = pedir_texto("  Nombre del autor: ")
            separador()
            biblioteca.libros_por_autor(autor)

        # â”€â”€ USUARIOS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

        elif opcion == "5":
            encabezado("REGISTRAR NUEVO USUARIO")
            id_usuario = pedir_texto("  ID del usuario (ej. U01): ")
            nombre     = pedir_texto("  Nombre completo         : ")
            separador()
            biblioteca.agregar_usuario(id_usuario, nombre)

        elif opcion == "6":
            encabezado("LISTADO DE USUARIOS")
            biblioteca.consultar_usuarios()

        # â”€â”€ PRÃ‰STAMOS â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

        elif opcion == "7":
            encabezado("PRESTAR LIBRO")
            isbn       = pedir_texto("  ISBN del libro   : ")
            id_usuario = pedir_texto("  ID del usuario   : ")
            separador()
            biblioteca.prestar_libro(isbn, id_usuario)

        elif opcion == "8":
            encabezado("DEVOLVER LIBRO")
            isbn       = pedir_texto("  ISBN del libro   : ")
            id_usuario = pedir_texto("  ID del usuario   : ")
            separador()
            biblioteca.devolver_libro(isbn, id_usuario)

        elif opcion == "9":
            encabezado("PRÃ‰STAMOS ACTIVOS")
            biblioteca.listar_prestamos_activos()

        # â”€â”€ REPORTES â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

        elif opcion == "10":
            encabezado("TOP 3 LIBROS MÃS PRESTADOS")
            biblioteca.top3_libros_mas_prestados()

        # â”€â”€ SALIR â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

        elif opcion == "0":
            separador("â•")
            print("  Hasta luego. El sistema se ha cerrado correctamente.")
            separador("â•")
            break

        else:
            print(f"  âš  La opciÃ³n '{opcion}' no es vÃ¡lida. Por favor elige un nÃºmero del 0 al 10.")

        # Pausa antes de volver al menÃº para que el usuario pueda leer el resultado
        input("\n  Presiona Enter para volver al menÃº principal...")
````
